```python
features = crop.columns.to_list()
features.remove('label')
print(features)
```

```python
#Visualize the distribution of each features

import seaborn as sns
fig,ax=plt.subplots(3,3,figsize=(15,15))
for i,subplot in zip(features,ax.flatten()):
    sns.histplot(crop[i],ax=subplot,kde=True)
    subplot.set_title(i)
plt.tight_layout()
plt.show()
```

```python
# PLOTING SCATTEER PLOT
fig,ax=plt.subplots(3,3,figsize=(15,15))
for i,subplot in zip(features,ax.flatten()):
    sns.histplot(x=i,y='label',data=crop,ax=subplot)
    subplot.set_title(i)
plt.tight_layout()
plt.show()
```

```python
# FINDING OUTLINERS - BOXPLOT
import seaborn as sns
fig,ax=plt.subplots(3,3,figsize=(15,15))
for i,subplot in zip(features,ax.flatten()):
    sns.boxplot(crop[i],ax=subplot)
plt.tight_layout()
plt.show()
```
